1. Install and import necessary packages

```python
import pandas as pd
import numpy as np
```

2. Load and check survey data

```python
# Load the dataset as a dataframe (set the first row as header)
df = pd.read_csv('Manisiesta_April 25, 2024_15.40.csv', header=1)
```

```python
# Drop the first row containing unnecessary information and reset the index
df = df[1:]
df.reset_index(drop=True, inplace=True)

# Display the dataframe to get first look
display(df)
```

```python
# Drop unnecessary columns
df = df.drop(['Start Date', 'End Date', 'Response Type', 'Finished', 'Response ID', 'Distribution Channel', 'User Language'], axis=1)
```

```python
# Convert data type for dataframe to string
df = df.astype('string')
df[['Progress', 'Duration (in seconds)']] = df[['Progress', 'Duration (in seconds)']].astype(int)
```

```python
# Function to extract year
def extract_year(date):
    if pd.isna(date):
        return np.nan
    elif '-' in date:
        return date.split('-')[-1]
    else:
        return date

# Apply function to the column
df["What is your year of birth?"] = df["What is your year of birth?"].apply(extract_year)
```

```python
# Drop rows where progress is less than 79 (Here they answered one of the manifesto stance questions, so might still be useful)
for index, row in df.iterrows():
    if row['Progress'] < 79:
        df.drop(index, inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Drop rows where duration is less than 60 seconds
for index, row in df.iterrows():
    if row['Duration (in seconds)'] < 60:
        df.drop(index, inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)
```
