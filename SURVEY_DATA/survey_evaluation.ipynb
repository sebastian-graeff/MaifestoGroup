```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Clean_Manisiesta_April 25, 2024_15.40.csv')
```

```python
display(df.head())
```

```python
# Plot distribution of 'year of birth'
plt.figure(figsize=(10, 6))
sns.histplot(df['What is your year of birth?'], bins=30, kde=True)
plt.title('Distribution of Year of Birth')
plt.xlabel('Year of Birth')
plt.ylabel('n survey participants')
plt.show()

# Plot distribution of 'gender'
df['How do you describe yourself? - Selected Choice'] = pd.Categorical(df['How do you describe yourself? - Selected Choice'], categories=['Male', 'Female', 'Non-binary / third gender', 'Prefer not to say'], ordered=True)
plt.figure(figsize=(10, 6))
sns.histplot(df['How do you describe yourself? - Selected Choice'])
plt.title('Distribution of gender')
plt.xlabel('gender')
plt.ylabel('n survey participants')
plt.show()


# Plot distribution of 'employment status'
plt.figure(figsize=(10, 6))
sns.histplot(df['What best describes your employment status over the last three months?'])
plt.title('Distribution of Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('n survey participants')
plt.show()

# Plot distribution of 'political orientation'
plt.figure(figsize=(10, 6))
sns.histplot(df['Some people talk about politics in terms of left, center and right. Where would you place yourself? - Political orientation'])
plt.title('Distribution of Political Orientation')
plt.xlabel('Political Orientation')
plt.ylabel('n survey participants')
plt.show()
```

```python
# Calculate the number of unique countries in the 'country' column
unique_countries_count = df['List of Countries'].nunique()
# Calculate the count of non-null entries in the 'other_countries' column
non_null_other_countries_count = df["What other countries' culture do you identify with?"].notnull().sum()
# Calculate the percentage of 'country' values that say 'Netherlands'
netherlands_count = (df['List of Countries'] == 'Netherlands').sum()
total_countries_count = df['List of Countries'].count()
netherlands_percentage = (netherlands_count / total_countries_count) * 100


print(f"Number of unique countries in the 'country' column: {unique_countries_count}")
print(f"Number of survey participants with another country identification: {non_null_other_countries_count}")
print(f"Percentage of countries that are 'Netherlands': {netherlands_percentage:.2f}%")

```
