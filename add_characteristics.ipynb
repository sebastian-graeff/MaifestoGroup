```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm

# Load data
data = pd.read_csv(r"CLEAN_DATA/Europe_2015_manifesto_gini_ps.csv")
welfare = pd.read_csv(r"RAW_DATA/social_welfare_spending.csv")

# Reduce dataframe to country, year and welfare spending and rename them
welfare = welfare[['Reference area', 'TIME_PERIOD', 'OBS_VALUE']]
welfare.rename(columns={'Reference area': 'countryname', 'TIME_PERIOD': 'year', 'OBS_VALUE': 'welfare spending'}, inplace=True)
```

```python
# extract the year from the edate column
data['year'] = pd.to_datetime(data['edate']).dt.year
# merge the two datasets
merged_df = pd.merge(data, welfare, on=['countryname', 'year'], how='left')
#reduce it to only 2015-2020
merged_df = merged_df[(merged_df['year'] >= 2015) & (merged_df['year'] <= 2020)]

display(merged_df.head())
```
