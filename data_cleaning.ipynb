```python
import pandas as pd
import numpy as np
import awoc #import the a-world-of-countries (awoc) library
```

2. Load and check dataset

```python
# Load the datasets as a dataframes
df_manifesto = pd.read_csv('RAW_DATA/MPDataset_MPDS2023a.csv')
df_gini = pd.read_csv('RAW_DATA/economic-inequality-gini-index.csv')
```

```python
# Display dataframes to get first look at data
display(df_manifesto)
display(df_gini)
```

```python
# Print a concise summary of the dataframe (determine datatype for each column)
df_manifesto.info(verbose=True)
df_gini.info(verbose=True)
```

3. Change datatypes where necessary

```python
# Change datatypes in df_manifesto for columns countryname, partyname, partyabbrev, candidatename, corpusversion, datasetversion, and id_perm from object into string.
columns_to_convert = ['countryname', 'partyname', 'partyabbrev', 'candidatename', 'corpusversion', 'datasetversion', 'id_perm']
df_manifesto[columns_to_convert] = df_manifesto[columns_to_convert].astype('string')
```

```python
# Change datatype in df_manifesto for edate from object into datetime
df_manifesto['edate'] = pd.to_datetime(df_manifesto['edate'], format='%d/%m/%Y')
```

```python
# Change datatypes in df_gini for columns Entity and Code from object to string
df_gini[['Entity', 'Code']] = df_gini[['Entity', 'Code']].astype('string')
```

```python
# Double check datatypes
df_manifesto.info(verbose=True)
df_gini.info(verbose=True)
```

4. Drop possible duplicate rows

```python
df_manifesto = df_manifesto.drop_duplicates()
df_gini = df_gini.drop_duplicates()
```

5. Save clean df_manifesto and df_gini seperately into csv files

```python
df_manifesto.to_csv("CLEAN_DATA/clean_MPDataset_MPDS2023a.csv")
df_gini.to_csv("CLEAN_DATA/clean_economic-inequality-gini-index.csv")
```

6. Combine datasets into one

6.1 Drop unnecessary columns from df_gini

```python
df_gini = df_gini.drop(columns=['Code'])
```

6.2 Relabel Entity and Year column in df_gini to match with df_manifesto

```python
df_gini = df_gini.rename(columns={"Entity" : "countryname", "Year" : "year"})
```

6.3 Extract year from edate in df_manifesto

```python
df_manifesto['year'] = df_manifesto['edate'].dt.year
```

6.4 Merge dataframes based on 'countryname' and 'year' columns

```python
df = pd.merge(df_manifesto, df_gini, on=['countryname', 'year'], how='left')
```

6.5 Drop year column (no longer needed)

```python
df = df.drop(columns=['year'])
```

6.6 Check dataset after merge

```python
display(df)
```

7. Save combined data to csv file

```python
df.to_csv("CLEAN_DATA/clean_manifesto_gini.csv")
```

8. Create a dataframe with elections from 2015 only

```python
# Filter rows where the year is 2015, saving in new dataframe
df_2015 = df[df['edate'].dt.year == 2015]
```

```python
display(df_2015)
```

9. Save 2015 data in separate csv file

```python
df_2015.to_csv("CLEAN_DATA/2015_manifesto_gini.csv")
```

10. Create a list of European countries and create a dataframe with only european countries

```python
# Initialize AWOC class
my_world = awoc.AWOC()

# Create list of European countries and print it
europe_countries = my_world.get_countries_list_of('Europe')
print(europe_countries)
```

```python
europe_df = df[df['countryname'].isin(europe_countries)]

display(europe_df)
```

11. Save Europe data in separate csv file

```python
europe_df.to_csv("CLEAN_DATA/Europe_manifesto_gini.csv")
```

12. Create dataframe with only european countries and with elections from 2015 only

```python
europe_2015_df = df_2015[df_2015['countryname'].isin(europe_countries)]

display(europe_2015_df)
```

13. Save 2015 Europe data in separate csv file

```python
europe_2015_df.to_csv("CLEAN_DATA/Europe_2015_manifesto_gini.csv")
```

```python
# Filter rows where the year is 2015 - 2020, saving in new dataframe
df_2015_2020 = df[(df['edate'].dt.year >= 2015) & (df['edate'].dt.year <= 2020)]

# Filter European countries
europe_2015_2020_df = df_2015_2020[df_2015_2020['countryname'].isin(europe_countries)]
display(europe_2015_2020_df)
```
