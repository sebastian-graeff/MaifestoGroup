```python
import pandas as pd
import numpy as np
import awoc #import the a-world-of-countries (awoc) library
```

2. Load and check dataset

```python
# Load the dataset as a dataframe
df = pd.read_csv('RAW_DATA/MPDataset_MPDS2023a.csv')
```

```python
# Display dataframe to determine data is in it
display(df)
```

```python
# Print a concise summary of the dataframe (determine datatype for each column)
df.info(verbose=True)
```

3. Change datatypes where necessary

```python
# Change datatype for columns countryname, partyname, partyabbrev, candidatename, corpusversion, datasetversion, and id_perm from object into string.
columns_to_convert = ['countryname', 'partyname', 'partyabbrev', 'candidatename', 'corpusversion', 'datasetversion', 'id_perm']
df[columns_to_convert] = df[columns_to_convert].astype('string')
```

```python
# Change datatype for edate from object into datetime
df['edate'] = pd.to_datetime(df['edate'], format='%d/%m/%Y')
```

```python
# Double check datatypes
df.info(verbose=True)
```

4. Drop possible duplicate rows

```python
df = df.drop_duplicates()
```

5. Save clean data to csv file

```python
df.to_csv("CLEAN_DATA/clean_MPDataset_MPDS2023a.csv")
```

6. Create a dataframe with elections from 2015 only

```python
# Filter rows where the year is 2015, saving in new dataframe
df_2015 = df[df['edate'].dt.year == 2015]
```

```python
display(df_2015)
```

7. Save 2015 data in separate csv file

```python
df.to_csv("CLEAN_DATA/2015_MPDataset_MPDS2023a.csv")
```

8. Create a list of European countries and create a dataframe with only european countries

```python
# Initialize AWOC class
my_world = awoc.AWOC()

# Create list of European countries and print it
europe_countries = my_world.get_countries_list_of('Europe')
print(europe_countries)
```

```python
europe_df = df[df['countryname'].isin(europe_countries)]

display(europe_df)
```

9. Save Europe data in separate csv file

```python
df.to_csv("CLEAN_DATA/Europe_MPDataset_MPDS2023a.csv")
```

10. Create dataframe with only european countries and with elections from 2015 only

```python
europe_2015_df = df_2015[df_2015['countryname'].isin(europe_countries)]

display(europe_2015_df)
```
